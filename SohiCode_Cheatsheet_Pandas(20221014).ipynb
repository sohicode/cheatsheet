## 데이터 전처리

```python
df = pd.DataFrame([['rex', 11, np.NaN, 35],
                   ['dino', 9, 125, 30], ['mew', 13, 145, np.NaN],
                   ['pengha', np.NaN, 115, 20], ['kiki', 8, 130, 35]],
                  index = ['411', '211', '611', '111', '112'],
                  columns = ['name', 'age', 'height', 'weight'])
```

```python
df.isna()
```

```python
df.isnull()
```

```python
df[df.weight.isnull()]
```

```python
df.notnull()
```

```python
df.dropna()
```

```python
df.weight.dropna()
```

```python
df.fillna(0)
```

```python
df['age'].fillna(df['age'].mean().astype(int))
```

```python
df.drop_duplicates()
```
